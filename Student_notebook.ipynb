## 2.1 Relabeling: Label encoding vs One hot encoding

The dataset has 17 numerical variables and 16 categorical variables. 

12 out of 16 categorical variables are binary, while the remaining 4 are nominal. 

The numerical variables are not to be touched and the binary values will be converted to 0 and 1. 

However, for the nominal variables we need to be **very** careful. Take for example variable number 9, "Mjob". It can take 5 different values: 1) Teacher, 2) Health, 3) Services, 4) at_home, 5) other.

The first naive idea is to assign integers from 0 to 4 to the five different values.

This straightforward technique is called **label encoding**. Applying it in this scenario would be shortsighted: we are implicitly stating that a teacher job is closer to an health job while being more different than staying at home. This is just an example but the same reasoning can be applied to all the combinations. The algorithms that I will be using after could be badly affected by this decision, especially in the regression analysis. 

In a nutshell, label encoding is useful when applied to values that can be ordedered or have some sort of heriarchy but this is not the case.

The problem can be solved by applying a more advanced technique: **one hot encoding**.
Each category value is converted into a new column: we assign 1 to the corresponding value and 0 to everything else.

We managed to solve the previous problem. However, with this dataset I am adding 17 new columns (variables): one for each possible value of the nominal variables. We also eliminate the 4 columns corresponding to the starting nominal variables. All in all, netting an addition of 13 new shiny columns to the 33 I was starting with, a **non** negligible amount of nearly **40%**. 

In this context, I do not care too much for two reasons: I still have $n >> p$ and I will be performing Principal component analysis (PCA) to greatly reduce the number of variables anyway.

However it is clear that this technique does not scale too well if there are many nominal variables with a lot of possible values. 


```python
# Relabeling and then hot encoding

#Creating instance
df_labelled = df
### I need to label encode before hot encoding the nominal variables, I need to label encode the binary values anyway

# LABEL ENCODING: all categorical variables are now number, binary -> 0 and 1 and nominal -> 0 1 2 3 ... k
labelencoder = LabelEncoder()
for cols in df.columns:
    if df_labelled[cols].dtypes == 'object':
        df_labelled[cols] = labelencoder.fit_transform(df_labelled[cols])

df_labelled.head(3)
# ONE HOT ENCODING: now I need to add cols as explained before. 
#I can use the feature of one hot encoding or use dummy variables, 
#the former is easier to use I just need to avoid drop_first=True

# Adding the new cols
df_labelled = pd.concat([df_labelled,pd.get_dummies(df_labelled['Mjob'], prefix='Mjob')],axis=1)
df_labelled = pd.concat([df_labelled,pd.get_dummies(df_labelled['Fjob'], prefix='Fjob')],axis=1)
df_labelled = pd.concat([df_labelled,pd.get_dummies(df_labelled['reason'], prefix='reason')],axis=1)
df_labelled = pd.concat([df_labelled,pd.get_dummies(df_labelled['guardian'], prefix='guardian')],axis=1)


# Removing the old nominal variables
df_labelled.drop(['Mjob'],axis=1, inplace=True)
df_labelled.drop(['Fjob'],axis=1, inplace=True)
df_labelled.drop(['reason'],axis=1, inplace=True)
df_labelled.drop(['guardian'],axis=1, inplace=True)
```

## 3.1 Principal component analysis (PCA)

Principal component analysis (PCA) is a statistical procedure that uses an orthogonal transformation to convert $n$ variables of a dataset into a new set of $k$ new variables which are linearly uncorrelated and called principal components. 
Variance is infomation, that's why the principal components are ranked according to the variance of data along them.  
Then, the idea behind this technique is to have $k < n$, while retaining the highest amount of variance. 
Mathematically, this is achieved by the following steps:
 1. Normalize X
 2. Find eigenvectors and eigenvalues of the corresponding correlation matrix, ranking them from highest eigenvalue to     lowest.
 3. Find W projection matrix whose k columns are the top k eigenvectors. 
 4. Find the $X_{new} = XW$.
 
$X_{new}$ is totally different because we are looking at the data from a different angle, that is from a different basis. This means we are losing interpretability with the process. The choice of $k$ is on me: I pick a number of eigenvectors that allow me to explain the data as best as possible without abusing the number of variables. 

```python
#I split the data in "X" and "Y":
df_y = df['G3']
df_x = df.drop(['G3'], axis=1)

#Normalize:
df_x_normalized = (df_x - df_x.mean()) / df_x.std()


# calculate the principal components
df_pca = PCA().fit(df_x_normalized)

#Plotting cumulative variance
plt.plot(np.cumsum(df_pca.explained_variance_ratio_))
plt.title('Cumulative variance')
plt.xlabel('Number of components')
plt.ylabel('Variance explained')
```

```python
cumvar = np.cumsum(df_pca.explained_variance_ratio_)
```

```python
cumvar[29]
```

I decide to use the first k = 30 components (from 0 to 29), because they retain nearly 90% of variance.

```python
n_used = 29
df_x_pca = np.dot(df_x_normalized.values, df_pca.components_[:n_used,:].T)
df_x_pca = pd.DataFrame(df_x_pca, columns=["PC#%d" % (x + 1) for x in range(n_used)])
```
